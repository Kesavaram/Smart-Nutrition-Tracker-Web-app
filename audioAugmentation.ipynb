```python
import numpy as np
import pandas as pd

from pydub import AudioSegment
import random

# Load the original sound file
original_sound = AudioSegment.from_file("input/trainingData/bacons0.wav")

# Define the augmentations you want to apply
def add_noise(sound):
    noise_level = -30
    noise = AudioSegment.silent(duration=len(sound), frame_rate=sound.frame_rate)
    noise = noise.apply_gain(noise_level)
    return sound.overlay(noise)

def change_speed(sound):
    speed_change = random.uniform(0.9, 1.1)
    return sound.speedup(playback_speed=speed_change)

def change_pitch(sound):
    pitch_change = random.uniform(-300, 300)
    return sound.shift_pitch(pitch_change)

# Generate the augmented sound files
augmented_sounds = []
for i in range(15):
    augmented_sound = original_sound
    augmented_sound = add_noise(augmented_sound)
    augmented_sound = change_speed(augmented_sound)
    #augmented_sound = change_pitch(augmented_sound)
    augmented_sounds.append(augmented_sound)
    out_file = ".//AugmentedFile//word{0}.wav".format(i)
    augmented_sound.export(f"augmented_{i}.wav", format="wav")
```
